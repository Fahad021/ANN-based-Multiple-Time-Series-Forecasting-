### data reshaping - experimental code

```python
def lagged_df(lag, df):
    col_names = ["lag{}".format(i) for i in range(1, lag+1)]
    lagged = []
    for i in range(1,lag+1):
        lagged.append(df.iloc[:,0].shift(i).values)
    return pd.DataFrame(lagged, col_names).T

def leading_df(lead, df):
    col_names = ["lead{}".format(i) for i in range(1, lead+1)]
    lead_data = []
    for i in range(1,lead+1):
        lead_data.append(df.iloc[:,0].shift(-1*i).values)
    return pd.DataFrame(lead_data, col_names).T
```

```python
dummy     = df[[df.columns[0]]]
lagged_df = lagged_df(lag=24*7, df=dummy)
lead_df   = leading_df(lead=72, df= dummy)
dummy     = dummy.reset_index(drop=True)
main_df = pd.concat([lagged_df,dummy,lead_df], axis=1)
main_df= main_df.dropna()
```

# final code for generating aggregated training data out of sampled premises

```python
df
```

```python
df.columns
```

```python
import numpy
from time import sleep
from tqdm import tqdm

list_of_dataframes = []
for premise_id in tqdm(df.columns):
    # create lag features and 72-hr ahead outputs columns
    dummy     = df[[premise_id]]
    dummy.columns = ['kwh']
    #lagged_df = lagged_df(lag = 24*7, dataframe=dummy)
    lag = 24*7
    col_names = ["lag{}".format(i) for i in range(1, lag+1)]
    lagged = []
    for i in range(1,lag+1):
        lagged.append(dummy.iloc[:,0].shift(i).values)
    lagged_df = pd.DataFrame(lagged, col_names).T
    
    #lead_df   = leading_df(lead=24*3, dataframe= dummy)
    lead = 24*3
    col_names = ["lead{}".format(i) for i in range(1, lead+1)]
    lead_data = []
    for i in range(1,lead+1):
        lead_data.append(dummy.iloc[:,0].shift(-1*i).values)
    lead_df = pd.DataFrame(lead_data, col_names).T
    
    dummy     = dummy.reset_index()
    
    # creating columns of 
    # lag(t= [(t-1)...(t-168)])+current (t= 0)+ future (t= [+1,...+72])
    main_df   = pd.concat([lagged_df,dummy,lead_df], axis=1)
    
    main_df   = main_df.dropna()
    
    #create premise id features
    a = numpy.full(shape= (main_df.shape[0], len(encoded_ids[premise_id].values)),
               fill_value = encoded_ids[premise_id].values)
    
    id_cols = ["encoded_id_{}".format(i) for i in range(0, len(encoded_ids[premise_id].values))]
    premise_id_df = pd.DataFrame(data= a, columns= id_cols)
    main_df = main_df.reset_index(drop=True)
    
    # add premise id columns
    main_df = pd.concat([premise_id_df, main_df], axis=1)
    
    #append
    list_of_dataframes.append(main_df)
```

```python
def lagged_df(lag, df):
    col_names = ["lag{}".format(i) for i in range(1, lag+1)]
    lagged = []
    for i in range(1,lag+1):
        lagged.append(df.iloc[:,0].shift(i).values)
    return pd.DataFrame(lagged, col_names).T

lagged_df = lagged_df(lag=24*7, df=dummy)
lagged_df
```

```python
def leading_df(lead, df):
    col_names = ["lead{}".format(i) for i in range(1, lead+1)]
    lead_data = []
    for i in range(1,lead+1):
        lead_data.append(df.iloc[:,0].shift(-1*i).values)
    return pd.DataFrame(lead_data, col_names).T

lead_df = leading_df(lead=72, df= dummy)
lead_df 
```

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import RepeatedKFold
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dense, Activation
from keras.callbacks import EarlyStopping
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import train_test_split, cross_val_score
 
# get the model
def get_model(n_inputs, n_outputs):
    model = Sequential()
    model.add(Dense(128, input_dim=n_inputs, kernel_initializer='he_uniform', activation='relu'))
    model.add(Dense(128, kernel_initializer='he_uniform' , activation= 'relu'))
    model.add(Dense(128, kernel_initializer='he_uniform' , activation= 'relu'))
    model.add(Dense(n_outputs))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
 
# evaluate a model using repeated k-fold cross-validation
def evaluate_model_kfold(X, y):
    results = list()
    n_inputs, n_outputs = X.shape[1], y.shape[1]
    # define evaluation procedure
    cv = RepeatedKFold(n_splits=5, n_repeats=3, random_state=1)
    # enumerate folds
    for train_ix, test_ix in cv.split(X):
        # prepare data
        X_train, X_test = X[train_ix], X[test_ix]
        y_train, y_test = y[train_ix], y[test_ix]
        # define model
        model = get_model(n_inputs, n_outputs)
        # fit model
        monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, 
                        patience=5, verbose=1, mode='auto', 
                        restore_best_weights=True)
        model.fit(X_train,y_train,validation_data=(X_test,y_test),
                  callbacks=[monitor],verbose=2,epochs=10)
        #model.fit(X_train, y_train, verbose=1, epochs=100)
        # evaluate model on test set
        mse = model.evaluate(X_test, y_test, verbose=0)
        results.append(mse)
    return results
 
    
# evaluate a model using repeated ts-crossvalidation
def evaluate_model_tscv(X, y):
    results = list()
    n_inputs, n_outputs = X.shape[1], y.shape[1]
    # define evaluation procedure
    tscv = TimeSeriesSplit(n_splits=5)
    # enumerate folds
    for train_ix, test_ix in tscv.split(X):
        # prepare data
        X_train, X_test = X[train_ix], X[test_ix]
        y_train, y_test = y[train_ix], y[test_ix]
        # define model
        model = get_model(n_inputs, n_outputs)
        # fit model
        monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, 
                        patience=5, verbose=1, mode='auto', 
                        restore_best_weights=True)
        model.fit(X_train,y_train,validation_data=(X_test,y_test),
                  callbacks=[monitor],verbose=2,epochs=0)
        # evaluate model on test set
        mse = model.evaluate(X_test, y_test, verbose=0)
        results.append(mse)
    return results
```

```python
# evaluate model
results = evaluate_model_kfold (X.values, y.values)
# summarize performance
print('MSE: %.3f (%.3f)' % (mean(results), std(results)))
```

```python
# evaluate model
results = evaluate_model_tscv(X.values, y.values)
# summarize performance
print('MSE: %.3f (%.3f)' % (mean(results), std(results)))
```
